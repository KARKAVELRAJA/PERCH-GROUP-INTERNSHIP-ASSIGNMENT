```python
# MEASURE OF SKEWNESS OF "bmi", "age" AND "charges" COLUMNS

for i in "bmi","age","charges":
  print(i,df[i].skew())
```

```python
# CHECKING THE PRESENCE OF OUTLIERS IN "bmi", "age" AND "charges" COLUMNS

# OUTLIERS IN "bmi" COLUMN

bmi_outliers = []

data1  = sorted(df["bmi"])

X1, X2 = np.percentile(data1, [25, 75])

IQR1   = X2 - X1

Lower_Bound1 = X1 - (1.5 * IQR1) 
Upper_Bound1 = X2 + (1.5 * IQR1) 


for i in data1:
     if i > Upper_Bound1 or i < Lower_Bound1:
          bmi_outliers.append(i)

print("NUMBER OF OUTLIERS IN 'bmi' COLUMN :",len(bmi_outliers))
print("They are :",bmi_outliers)


print("\n") 



# OUTLIERS IN "age" COLUMN

age_outliers = []

data2  = sorted(df["age"])

Y1, Y2 = np.percentile(data2, [25, 75])

IQR2   = Y2 - Y1

Lower_Bound2 = Y1 - (1.5 * IQR2) 
Upper_Bound2 = Y2 + (1.5 * IQR2) 

for j in data2:
     if j > Upper_Bound2 or j < Lower_Bound2:
          age_outliers.append(j)

print("NUMBER OF OUTLIERS IN 'age' COLUMN :",len(age_outliers))
print("They are :",age_outliers)


print("\n")



# OUTLIERS IN "charges" COLUMN

charges_outliers = []

data3  = sorted(df["charges"])

Z1, Z2 = np.percentile(data3, [25, 75])

IQR3   = Z2 - Z1

Lower_Bound3 = Z1 - (1.5 * IQR3) 
Upper_Bound3 = Z2 + (1.5 * IQR3) 

for k in data3:
     if k > Upper_Bound3 or k < Lower_Bound3:
          charges_outliers.append(k)

print("NUMBER OF OUTLIERS IN 'charges' COLUMN :",len(charges_outliers))
print("They are :",charges_outliers)

```

##**ANSWERING THE QUESTIONS WITH STATISTICAL EVIDENCE**

### **DO CHARGES OF PEOPLE WHO SMOKE DIFFER SIGNIFICANTLY FROM THE PEOPLE WHO DON'T SMOKE?**

```python
df.groupby("smoker").max()["charges"]
```

**Yes, The charges of people who smoke differ significantly from the people who don't smoke.**

### **DOES BMI OF MALES DIFFER SIGNIFICANTLY FROM THAT OF FEMALES ?**

```python
df.groupby("sex").max()["bmi"]
```

**So, BMI value of Males are not that much differ from that of Females**

### **IS THE PROPORTION OF SMOKERS SIGNIFICANTLY DIFFERENT IN DIFFERENT GENDERS ?**

```python
result = []

df_smoker_yes = df[df["smoker"]=="yes"]

for i in df_smoker_yes["sex"].unique():
  result.append([i,len(df_smoker_yes[df_smoker_yes["sex"]==i])])

pd.DataFrame(result,columns=["SEX","NUMBER OF SMOKERS"])
```

**So, the Proportion of Smokers are not that much different with different genders**

### **IS THE DISTRIBUTION OF BMI ACROSS WOMEN WITH NO CHILDREN, ONE CHILD AND TWO CHILDREN, THE SAME ?**

```python
df_women = df[df["sex"]=="female"]

df_women_0_children = df_women[df_women["children"]==0]
df_women_1_children = df_women[df_women["children"]==1]
df_women_2_children = df_women[df_women["children"]==2]


sns.displot(data=df_women_0_children,x="bmi",kind="hist")
sns.displot(data=df_women_1_children,x="bmi",kind="hist")
sns.displot(data=df_women_2_children,x="bmi",kind="hist")
```

**Thus, Distribution of BMI across Women with NO children, ONE child and TWO children are nearly same**
